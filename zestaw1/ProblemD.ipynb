```python
import numpy as np
from Ruins import *
import matplotlib.pyplot as plt
```

```python
def mean(gameInfo : list):
    suma = 0

    for el in gameInfo:
        suma += el.roundsNumber

    return suma // len(gameInfo)
```

```python
def getRoundsNumber(gameInfo : list):
    return np.array([single_game.roundsNumber for single_game in gameInfo])
```

```python
aWinProb = [0.5, 0.2, 0.8]
startCapital = [[50, 50]]
N = 20000
gameInfo = []
```

```python
def game(a , b, probA, N):
    gameInfo = []

    for i in range(N):
        playerA = State(a, probA)
        playerB = State(b, 1.0 - probA)
        gameStatistics = GameStatistics()
        game = Game(playerA, playerB, gameStatistics)
        game.play()
        gameInfo.append(game.gameStatistics)
    return gameInfo
```

```python
gameInfo =  game(50 , 50, 0.5, 20000)

print("For capital a: ", 50, " b: ", 50, " and probability of a wins: ", 0.5)
print("Mean length is ", mean(gameInfo))

plt.hist(getRoundsNumber(gameInfo), density=True, bins=25)
plt.show()
```
