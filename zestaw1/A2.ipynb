```python
import numpy as np

def cauchy(x0, gamma, size=1):
    
    rnd_u = np.random.uniform(0,1, size)
    
    return x0 + gamma * np.tan(np.pi * (rnd_u - 0.5))
```

```python
import matplotlib.pyplot as plt


def plot(x0, gamma):
    assert(gamma > 0)
    x = cauchy(x0, gamma, 100)

    plt.hist(x, bins=35, density=True)

    x_range = np.linspace(-50, 50, 500)
    
    y = 1 / (np.pi * (gamma + (x_range - x0)**2))
    plt.plot(x_range, y)
    
    
    print("Mean ", np.mean(x))
    print("Variance ", np.var(x))
    plt.show()
```

```python
plot(0, 1)
```

```python
plot(5,2)
```
