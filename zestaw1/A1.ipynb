```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def generate_points(N):
    points = []
    for i in range(N):
        point = np.random.uniform(-1.0, 1.0, 2)
        norm = np.linalg.norm(point)
        if norm < 1:
            points.append(point)
    return np.array(points)
    
```

```python
def transform_points(points):
    y1 = []
    y2 = []
    for p in points:
        v1 = p[0]
        v2 = p[1]
        R = np.linalg.norm(p)
        y1.append(np.sqrt((-2)*np.log(R**2))*(v1/R))
        y2.append(np.sqrt((-2)*np.log(R**2))*(v2/R))
    return np.array(y1), np.array(y2)
```

```python
def plot_pdf():
    x = np.linspace(-4, 4, 1000)
    p = norm.pdf(x, 0, 1)

    return plt.plot(x, p, 'k', linewidth=2)
```

```python
points = generate_points(100000)
```

```python
y1, y2 = transform_points(points)
```

```python
plt.hist(y1, density=True, bins=25)
plot_pdf()
plt.show()
```

```python
print("Mean: ", np.mean(y1), " Variance: ", np.var(y1))
```

```python
plt.hist(y2, density=True, bins=25)
plot_pdf()
plt.show()
```

```python
print("Mean: ", np.mean(y2), " Variance: ", np.var(y2))
```
