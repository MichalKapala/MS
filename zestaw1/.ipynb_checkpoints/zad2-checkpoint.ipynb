```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt 
from scipy import stats
from scipy.stats import norm
```

```python
def generate_and_plot(Y0,Y):
    fy=[]
    x = np.random.uniform(-15,15,200)
#     x.sort()
    for i in range(0,len(x)):
        #formula for inverse distribution
        value = Y0+Y*np.tan(np.pi*(x[i]-1/2))
        fy.append(value)
    plt.hist(fy,bins=90,density=True)
    x0 = np.linspace(-50,50,500)
    plt.plot(x0,1/(np.pi*Y*(1+((x0-Y0)/Y)**2)),color='black')
    print("Mean: ", np.mean(fy))
    print("Variance: ",np.var(fy))
    plt.show()
```

```python
generate_and_plot(0,0.5)
```

```python
generate_and_plot(-10,0.5)
```

```python
generate_and_plot(10,1)
```
